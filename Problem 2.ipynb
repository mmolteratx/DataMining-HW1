```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def distr(n):
    i = n

    arr = np.zeros(n)

    while(i > 0):

        s = np.random.binomial(1, 0.5, n)
        s = np.where(s < 0.5, -1, 1)

        a = np.sum(s) * (1 / math.sqrt(n))

        arr[n - i] = a
        i -= 1

    plt.hist(arr)
    plt.show()
```

```python
n = 5
distr(n)
```

```python
n = 50
distr(n)
```

```python
n = 250
distr(n)
```
