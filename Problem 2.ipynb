```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def z_n(n):
    arr = np.random.choice([-1,1], n)
    zn = arr.sum() / np.sqrt(n)
    return zn
```

```python
n = [5,10,50,250,1000]
fig, axs = plt.subplots(1,5, figsize=(15, 3), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)

axs = axs.ravel()

for i in range(5):
    data = np.array([z_n(n[i]) for _ in range(1000)])
    axs[i].hist(data, color='g')
    axs[i].set_title("n={}".format(n[i]))
    plt.tight_layout(pad=3.0)
```
